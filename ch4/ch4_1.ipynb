```python
import pandas as pd
import re

data4 = pd.read_csv('../data/订单表2022-4.csv',encoding='gbk')
data5 = pd.read_csv('../data/订单表2022-5.csv',encoding='gbk')
data6 = pd.read_csv('../data/订单表2022-6.csv',encoding='gbk')
data7 = pd.read_csv('../data/订单表2022-7.csv',encoding='gbk')
data8 = pd.read_csv('../data/订单表2022-8.csv',encoding='gbk')
data9 = pd.read_csv('../data/订单表2022-9.csv',encoding='gbk')

print("data4的行列数为：",data4.shape)
print("data5的行列数为：",data5.shape)
print("data6的行列数为：",data6.shape)
print("data7的行列数为：",data7.shape)
print("data8的行列数为：",data8.shape)
print("data9的行列数为：",data9.shape)

data=pd.concat([data4,data5,data6,data7,data8,data9],axis=0,ignore_index=True)
print("合并后的data的行列数为：",data.shape)

print("各属性缺失值为：\n",data.isnull().sum())

data.dropna(how='any',inplace=True)
print("删除后的data的行列数为：",data.shape)

data['市']=data['省市区'].str[3:6]

# 定义处理函数
def clean_from_x(text):
    # 从后面找到第一个 'X' 或 'x'，并删除该字符及其后的所有字符
    match = re.search(r'[Xx]', text[::-1])  # 反转字符串来从后向前查找
    if match:
        # 找到 'X' 或 'x' 的位置
        x_pos = len(text) - match.start() - 1  # 获取 'X' 或 'x' 在原字符串中的位置
        # 截取到 'X' 或 'x' 之前的内容
        text = text[:x_pos]
        
        # 检查 'X' 或 'x' 前面第二个位置是否为空格，删除该空格及其前面的字符
        if x_pos >= 2 and text[x_pos - 2] == ' ':
            # 删除空格及其前一个字符，再删除 'X' 或 'x'
            text = text[:x_pos - 2] + text[x_pos:]  # 删除空格和它前面的字符

    return text

# 对 '商品详情' 列进行处理
data['商品详情'] = data['商品详情'].apply(clean_from_x)

error_str = list(' ')
for i in error_str:
    data['商品详情'] = data['商品详情'].str.replace(i, '', regex=False)

data['商品名称']=data['商品详情']

print("现有属性为：\n",data.columns)
print("data的前5行为：\n",data.head())

data.to_csv('cleaned_data.csv', index=False)
```
